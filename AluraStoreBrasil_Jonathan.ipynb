### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head(10)
# loja4.describe()
# loja['Preço'].unique()
```

#1. Análise do faturamento



```python
lj1_preco = loja['Preço'].sum()
lj2_preco = loja2['Preço'].sum()
lj3_preco = loja3['Preço'].sum()
lj4_preco = loja4['Preço'].sum()
```

```python
print(f'O faturamento das lojas foram: \nLoja 1: {lj1_preco} \nLoja 2: {lj2_preco} \nLoja 3: {lj3_preco}  \nLoja 4: {lj4_preco}')
```

# 2. Vendas por Categoria

```python
lj1_venda_categoria = loja.groupby('Categoria do Produto').agg(Qtde_Produto=('Produto', 'count'), Preço_Produto=('Preço', 'sum')).sort_values('Qtde_Produto', ascending= False)
lj2_venda_categoria = loja2.groupby('Categoria do Produto').agg(Qtde_Produto=('Produto', 'count'), Preço_Produto=('Preço', 'sum')).sort_values('Qtde_Produto', ascending= False)
lj3_venda_categoria = loja3.groupby('Categoria do Produto').agg(Qtde_Produto=('Produto', 'count'), Preço_Produto=('Preço', 'sum')).sort_values('Qtde_Produto', ascending= False)
lj4_venda_categoria = loja4.groupby('Categoria do Produto').agg(Qtde_Produto=('Produto', 'count'), Preço_Produto=('Preço', 'sum')).sort_values('Qtde_Produto', ascending= False)
```

# 3. Média de Avaliação das Lojas


```python
media_nota_loja1 = loja['Avaliação da compra'].mean().round(2)
media_nota_loja2 = loja2['Avaliação da compra'].mean().round(2)
media_nota_loja3 = loja3['Avaliação da compra'].mean().round(2)
media_nota_loja4 = loja4['Avaliação da compra'].mean().round(2)
```

```python
print(f'A médias de nota das avaliações das lojas foram: \nLoja 1: {media_nota_loja1} \nLoja 2: {media_nota_loja2} \nLoja 3: {media_nota_loja3}  \nLoja 4: {media_nota_loja4}')
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_loja1 = loja['Produto'].value_counts()
mais_vendido_loja = produtos_loja1.index[0]
qtd_mais_vendido_loja = produtos_loja1.iloc[0]
menos_vendido_loja = produtos_loja1.index[-1]
qtd_menos_vendido_loja = produtos_loja1.iloc[-1]
######
produtos_loja2 = loja2['Produto'].value_counts()
mais_vendido_loja2 = produtos_loja2.index[0]
qtd_mais_vendido_loja2 = produtos_loja2.iloc[0]
menos_vendido_loja2 = produtos_loja2.index[-1]
qtd_menos_vendido_loja2 = produtos_loja2.iloc[-1]
######
produtos_loja3 = loja3['Produto'].value_counts()
mais_vendido_loja3 = produtos_loja3.index[0]
qtd_mais_vendido_loja3 = produtos_loja3.iloc[0]
menos_vendido_loja3 = produtos_loja3.index[-1]
qtd_menos_vendido_loja3 = produtos_loja3.iloc[-1]
######
produtos_loja4 = loja4['Produto'].value_counts()
mais_vendido_loja4 = produtos_loja4.index[0]
qtd_mais_vendido_loja4 = produtos_loja4.iloc[0]
menos_vendido_loja4 = produtos_loja4.index[-1]
qtd_menos_vendido_loja4 = produtos_loja4.iloc[-1]
```

```python
print(f'O produto mais vendido da Loja 1 foi o produto: {mais_vendido_loja} contendo {qtd_mais_vendido_loja} unidades e o menos vendido foi o produto: {menos_vendido_loja} contendo {qtd_menos_vendido_loja} unidades.')
print(f'O produto mais vendido da Loja 2 foi o produto: {mais_vendido_loja2} contendo {qtd_mais_vendido_loja2} unidades e o menos vendido foi o produto: {menos_vendido_loja2} contendo {qtd_menos_vendido_loja2} unidades.')
print(f'O produto mais vendido da Loja 3 foi o produto: {mais_vendido_loja3} contendo {qtd_mais_vendido_loja3} unidades e o menos vendido foi o produto: {menos_vendido_loja3} contendo {qtd_menos_vendido_loja3} unidades.')
print(f'O produto mais vendido da Loja 4 foi o produto: {mais_vendido_loja4} contendo {qtd_mais_vendido_loja4} unidades e o menos vendido foi o produto: {menos_vendido_loja4} contendo {qtd_menos_vendido_loja4} unidades.')
```

# 5. Frete Médio por Loja



```python
frete_loja1 = loja['Frete'].mean()
frete_loja2 = loja2['Frete'].mean()
frete_loja3 = loja3['Frete'].mean()
frete_loja4 = loja4['Frete'].mean()
```

```python
print(f'O frete médio das lojas foram: \nLoja 1: {frete_loja1} \nLoja 2: {frete_loja2} \nLoja 3: {frete_loja3}  \nLoja 4: {frete_loja4}')
```

# 6. Gráficos

```python
notas = {
    'Loja 1': media_nota_loja1,
    'Loja 2': media_nota_loja2,
    'Loja 3': media_nota_loja3,
    'Loja 4': media_nota_loja4
}

notas_series = pd.Series(notas)

notas_series.plot(
    kind='bar',
    color='blue',
    figsize=(8,5)
)

plt.title('Média de notas por lojas')
plt.xlabel('Lojas')
plt.ylabel('Média de notas')
plt.show()
```

```python
faturamento = {
    'Loja 1': lj1_preco,
    'Loja 2': lj2_preco,
    'Loja 3': lj3_preco,
    'Loja 4': lj4_preco
}
faturamento_series = pd.Series(faturamento)

faturamento_series.plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(7, 7),
    startangle=90,
    colormap='Set3',
    legend=False,
    ylabel=''
)

plt.title('% de faturamento total por loja')
plt.show()
```

```python
produtos = {
   f'Loja 1 - {mais_vendido_loja}': qtd_mais_vendido_loja,
   f'Loja 2 - {mais_vendido_loja2}': qtd_mais_vendido_loja2,
    f'Loja 3 - {mais_vendido_loja3}': qtd_mais_vendido_loja3,
    f'Loja 4 - {mais_vendido_loja4}': qtd_mais_vendido_loja4
}

produtos_series = pd.Series(produtos)

produtos_series.plot(
    kind='barh',
    color='purple',
    figsize=(8,5)
)

plt.title('Produtos mais vendidos por lojas')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Produtos')
plt.show()
```
